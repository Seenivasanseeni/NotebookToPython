```python
import torch
import torchvision
from torch.autograd import Variable
```

```python
#make test data
import numpy as np
x=np.arange(100,dtype=np.float32)
y=2*x+3
```

```python
x=(x-x.mean())/x.std()
y=(y-y.mean())/y.std()
print(x,y,sep="\n")
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.layer1=nn.Linear(1,3)
        self.layer2=nn.Linear(3,1) 
    def forward(self,x):
        x=self.layer1(x)
        x=self.layer2(x)
        return x
```

```python
network=Net()
opt=optim.SGD(network.parameters(),lr=0.01,momentum=0.1)
```

```python
def train():
    global x,y
    for i in range(1000):
        network.zero_grad()
        outputs=network(torch.from_numpy(x).view([-1,1]))
        loss=F.mse_loss(outputs,torch.from_numpy(y))
        print(loss)
        loss.backward()
        opt.step()
train()
```
